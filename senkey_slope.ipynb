```python
import pandas as pd
import numpy as np
```

# Loading data

```python
df = pd.read_excel("raw_data.xlsx")
```

# Calculate slope line using Sigmoid function

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
country = []
oldRank = []
newRank = []
size = []
t = []
excel_sig = []
excel_cur = []
color = []

texcel_sig = 1/df.shape[0]
itt = 12/(df.shape[0]-1)


for index , row in df.iterrows():
    tt = -6
    
    for i in range(df.shape[0]-1):
        country.append(row['Country'])
        oldRank.append(row['Rank 2013'])
        newRank.append(row['Rank 2015'])
        size.append(abs(row['Rank 2015']-row['Rank 2013']))
        color.append(row['Rank 2015']-row['Rank 2013'])
        
        t.append(tt)
        texcel_sig = sigmoid(tt)
        
        excel_sig.append(texcel_sig)
        excel_cur.append(row['Rank 2013'] + ((row['Rank 2015']-row['Rank 2013'])* texcel_sig))
        
        tt = tt + itt
        
```

```python
dftab = pd.DataFrame({'Country': country, '2013 Ranking': oldRank, '2015 Ranking':newRank,
                        'Excel Size': size, 'Excel Sigmoid':excel_sig, 't': t,
                        'Excel Calculated Curve': excel_cur, 'color':color
                       })
```

```python
dftab
```

```python
dftab.to_csv("senkey_slope_data.csv", index=False)
```
